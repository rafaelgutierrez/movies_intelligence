# Search tweets and create a word cloud of the text

### Set up a client for the Search API

```python
import tweepy
import json
import pandas as pd

 
# Change the xxx values with your keys
consumer_key = 'xxx'
consumer_secret = 'xxx'
access_token = 'xxx'
access_secret = 'xxx'
bearer_token='xxx'


client = tweepy.Client(
    wait_on_rate_limit = True,
    consumer_key = consumer_key,
    consumer_secret = consumer_secret,
    access_token = access_token,
    access_token_secret = access_secret,
    bearer_token = bearer_token,
)
```

### Retrieve from the Search API

```python
search = client.search_recent_tweets(
    # Query
    query = "coronavirus",
    tweet_fields = ["attachments", "author_id", "conversation_id", "created_at", 
                    "entities", "geo", "id", "in_reply_to_user_id", "lang", 
                    "public_metrics", "possibly_sensitive", "referenced_tweets", 
                    "reply_settings", "source", "text", "withheld"],
    max_results = 100
)

search_df = pd.DataFrame()
for i in search.data:
  temp_data = pd.json_normalize( i.data ,  sep = "_")
  search_df = search_df.append(temp_data, ignore_index=True)
search_df
```

# Streaming (realtime) tweets and create a word cloud of the text

First, you will need to apply for Academic Research access https://developer.twitter.com/en/products/twitter-api/academic-research

```python
import time

class my_streamer_class(tweepy.Stream):
    def on_status(self, status):
      print(status.id,"\t", status.text)  
      statuses_container.append(status)  

stream = my_streamer_class(
    consumer_key = consumer_key,
    consumer_secret = consumer_secret,
    access_token = access_token,
    access_token_secret = access_secret,
)
```

```python
statuses_container = []

stream.filter(languages=["en", "es"], track=["coronavirus"], threaded=True)
time.sleep(10)
stream.disconnect()
```

```python
streaming_df = pd.DataFrame()
for i in statuses_container:
  temp_data = pd.json_normalize( i._json ,  sep = "_")
  streaming_df = streaming_df.append(temp_data, ignore_index=True)
streaming_df
```
